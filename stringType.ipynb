```python
# 문자열의 길이를 구하는 len() 함수가 있다.
# 입력된 주민번호 글자수를 알아내는 코드
# id/pw에 대한 글자수 검증용으로 활용할 수 있다.
# 주민번호 입력할 때 Ssn 을 많이 사용함.

inputSsn = "120604-2345678"
type(inputSsn)
```

```python
len(inputSsn)  # - 도 포함해서 글자수 나옴
```

```python
if len(inputSsn)==14 :
    print("주민번호 14글자가 잘 입력 되었습니다.") #참일때(true)
else:
    print("주민번호 14글자가 잘못 입력 되었습니다.") #거짓일때(false)
```

```python
# 문자열 인덱싱은 문자를 표로 만들어 0번부터 주소를 넣은 값
# 문자열 슬라이싱은 문자를 인덱스로 잘라 가져옴.
word = "Life is too short, You need python"
#       0123456789                       33 번의 인덱싱을 가지고 있음
```

```python
word[-1] #인덱스의 -값은 맨 뒤에서부터
```

```python
# 주민번호 앞 2자리를 생년으로 추출 year
# 주민번호 3~4 자리를 생월로 추출 month
# 주민번호 5~6 자리를 생일로 추출 day
# 주민번호 8번째 1자리를 남여 추출하는 sex

year = inputSsn[0:2]
month = inputSsn[2:4]
day = inputSsn[4:6]
sex = inputSsn[7]

print("귀하의 생년은 " + year + "년")
print("귀하의 생년은 " + month + "월")
print("귀하의 생일은 " + day + "일 입니다.")

# 주민번호를 이용해서 남여 구분을 위한 코드
# 문자열로 인식된 숫자를 숫자로 계산시 타입변환 필요
ssnNumber = inputSsn[7]
# print(ssnNumber)
# type(ssnNumber)
ssnNumber = int(ssnNumber) # 문자가 숫자로 타입변환진행 int(문자) 식을 입력하면 문자열을 숫자로 바꿔줌
# type(ssnNumber)

if ssnNumber % 2 == 1 : # 숫자로된 성별을 2로 나눈 나머지가 1
    sex = "남자"
else:
    sex = "여자"
print(sex)
```
