```python
# 딕셔너리 타입은 key : value로 이루어진다.
# {key1 : value1, key2 : value2 ....}
# 대부분 웹 프로그램 언어는 json이라는 값으로 처리하는데 유사한 형태임.

dic = {"name" : "kkw", "age" : 40}
#        key    value   key    value

type(dic)
print(dic["name"])
#           key
```

```python
a = {1 : 'hi'}
print(a[1])
#       key
```

```python
b = {"a" : [1,2,3]}
#     k       v
print(b["a"])
```

```python
# 여기서 2를 출력하고 싶을때
print(b["a"][1]) # 이럴때는 주소를 쓸 수 있음.
```

```python
print(dic)
```

```python
# dic 변수에 email key를 갖는 값을 추가하는 방법
dic['email'] = "mbc@kkw.com"  # email값을 뒤에 추가함
print(dic)
```

```python
# dic 변수에 age를 삭제해보자.
del dic["age"]   # 삭제할 때 del 명령어 사용
print(dic)
```

```python
# dic 변수에 name을 바꿔보자.
dic['name'] = "김기원"
print(dic)
```

```python
# key를 리스트를 생성하려면 a.keys()
dic.keys()
```

```python
# key의 값들을 리스트로 생성하려면
dic.values()
```

```python
# key와 value의 쌍으로 튜플처리 하려면
dic.items()
```

```python
# get을 이용해 키를 넣으면 값이 나오도록
dic.get("name")
```

```python
# key에 해당하는 값을 가져올때 없는 키를 넣으면 에러발생
dic["age"]  #에러가 뜨는 이유는 없는 키를 가져왔기 때문.(KeyError)
# 이것은 프로그램 실행 중 치명적인 오류로 강제종료됨.
```

```python
dic.get("age","값없음")
#        key    key가 없을때 출력문
# .get을 쓰면 없는 값을 써도 오류가 발생하지 않음.
# 뒤에 나올 데이터 값을 입력해주면 그 값으로 출력됨.
```

```python
print(dic)
```

```python
"name" in dic  # dic 안에 name이 있는지 물어보는 것임.
# in 명령어는 리스트나, 튜플, 딕셔너리 안에 값이 있냐?
```

```python
"김기원" in dic  # 값이 아닌 key만 찾아옴
```

```python
# key를 이용해서 value - pop (값을 꺼내와 변수에 넣고 삭제한다)
print(dic)
```

```python
email = dic.pop("email")  # 변수에 넣고 삭제할 수 있음.
print(email)
```

```python
print(dic)
```

```python
age = dic.pop("age", "찾는 값이 없습니다.")
print(age)
```
