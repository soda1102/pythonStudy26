```python
# 파일로 더미데이터 및 추가자료를 관리해본다.
# 지금까지는 변수 선언 시 초기값을 리스트에 담아 제공받음.
# 그러나 프로그램을 종료하면 추가한 값은 사라짐.
# 자료를 영구적으로 보관하는 방법은 파일로 저장 또는 데이터베이스에 보관
# 자료를 보관하는 곳이 따로있음. 이는 서버라고 명칭함.

# 서버는 24시간 365일동안 멈춤없이 서비스를 제공하는 컴퓨터임.(자료보관 등)
# 서비스 : 웹서비스, 게임서비스, 쇼핑서비스, 은행뱅킹 서비스 등(제공자의 뜻도 있음)
# 클라이언트 : 제공을 받는 자

# 파일 저장 : 텍스트 파일(메모장), csv(엑셀) 파일, json(웹api) 등으로 저장 가능
```

```python
f = open("새파일.txt","w")
# open = 파일을 열 때 사용함.
# 옵션 : r -> read(읽기전용)
#       w -> writer(쓰기 가능)
#       a -> append(추가가능, 마지막행에 추가)
# 파일을 열 때 경로를 지정하지 않으면 같은 위치에 파일을 찾거나 쓴다.
f.close()
# 파일은 윈도우 정책에서 단일 처리를 함.(열고 작업이 끝나면 닫고 나와야함.)
# 안닫으면 다른 프로그램에서 열 수 없음!
```

```python
f = open("c:/mbc/새파일.txt","w")
f.close()
# 주의사항 : 경로에 / 를 붙이는게 일반적이고
# 만약 \를 사용하고 싶으면 \\ (2개) 사용하기
# ex. c:\\mbc\\새파일.txt
# \t 탭, \n 엔터, \는 이스케이프 문자로 기능이 있는 문자열
# 따라서 한번 쓰면 기능이 있는 문자열로 인식하기 때문에 두번써야함.
```

```python
f = open("c:/mbc/새파일.txt","w")
for i in range(1,11) :   # range는 1~10까지 값을 가지고 있다는 뜻
    data = "%d번째 줄입니다. \n" % i   # %d는 10진수
    #       정수처리        엔터   for i
    f.write(data)   #write() 기록!
    # 파일에 저장한다는 뜻

f.close()
```

```python
# 파일에 있는 내용을 가져와보자
# .readline() 파일에 있는 내용 줄단위(행 : 가로)로 가져오기
f = open("c:/mbc/새파일.txt","r")
line = f.readline()  # readline 기본값이 첫줄 가져오기
print(line)
f.close()
```

```python
# 모든 줄을 읽어 화면에 출력하고 싶으면
f = open("c:/mbc/새파일.txt","r")
while True:   # False처리 될때까지 반복문
    line = f.readline()   # 한줄씩 읽는다.
    if not line:
        break    # line이 없으면 break 중지 처리(암기하기)
    else:
        print(line)
f.close()
# 결과값에 엔터가 하나씩 더 있는 이유는
# 메모장 내용에 엔터가 적용이 되어있는 상태기 때문
# \n \r 라인피드 캐리지리턴(메모장에 기록이 없는 상태로 적용되어있음)
```

```python
# .readlines()  메모장 파일을 리스트로 한번에 불러옴.
# 파일의 모든 줄을 읽어 리스트 형태로 반환
# 반환값은 문자열 리스트 ['첫번째줄\n','두번째줄\n',,,]
# 파일 내용을 모두 메모리에 로드하기 때문에 파일 크기가 클 경우 메모리 부족!
# 장점 : 파일의 모든 줄을 한번에 가져와 처리해야할 때 편리

f = open("c:/mbc/새파일.txt","r")
lines = f.readlines()   # 파일에 있는 모든 내용을 리스트 처리함
#print(lines)
for lines in lines: print(lines)
f.close()
```

```python
# 메모장에 엔터효과인 \n 을 지우고 싶으면 strip() 사용
# strip() : 문자열 내에서 원하는 문자열 또는 공백을 제거함.
# lstrip() : 원하는 문자열 또는 공백을 왼쪽에서 제거
# rstrip() : 원하는 문자열 또는 공백을 오른쪽에서 제거
f = open("c:/mbc/새파일.txt","r")
lines = f.readlines()
for line in lines:
    line = line.strip()
    print(line)
f.close()
```

```python
# read()를 이용하여 파일 문자열 읽기
# 파일의 내용 전체를 문자열로 반환
f = open("c:/mbc/새파일.txt","r")
data = f.read()   #read()는 메모장 내용을 통째로 가져와서 읽는다.
print(data)   #for문을 사용하지 않고 전체를 읽어 출력
f.close()
```

```python
# for를 이용해 줄단위로 가져오기
f = open("c:/mbc/새파일.txt","r")
# 파일 객체는 기본적으로 for문과 같이 함께 사용하여 줄단위로 읽기도 가능
for line in f :
    print(line)
f.close()
```

```python
# 파일에 새로운 내용을 추가 append -> a
# w로 파일을 열면 이미 존재하는 파일이면 그 내용이 사라지고 덮어쓰기가 됨.
# a모드는 원래 있던 값을 유지하면서 새로운 값을 맨 아래에 추가
f = open("c:/mbc/새파일.txt","a")  #append(추가)
f.write("------------------")
for i in range(11,20):
    data = "%d번째 줄입니다. \n" % i
    f.write(data)
f.close()
```
