```python
# 리스트란 표 형식의 배열을 말한다.
# id = ["kkw" , "lhj" , "ksb"]
# pw = ["1234", "4321", "4567"]
# names = ["김기원", "임효정", "김수빈"] (이름은 여러명이기 때문에 s를 붙인다.)
# 리스트를 만들때는 대괄호[]로 감싸주고 쉼표, 로 구분한다.

# ex. a = [] 빈 리스트를 생성한다.  -> a = list()
#     b = [1, 2, 3] 정수를 입력할 수 있다.  -> b[1] => 2
#     c = [1, 2, 'kkw', 'lhj']와 같이 혼용하여 넣을 수 있다. (자바에서는 사용이 안됨)
#         학년 반 이름      id     -> 여러개를 섞을 수 있는걸 파이썬만 가능함. Java는 안됨.
#     d = [1, 2, ['kkw','1234']]  -> 2차원 리스트(배열 안에 또 배열)
#                   0      1
#          0  1        2
# d 변수에서 kkw를 가져오고 싶으면? d[2][0]
# f = [1,2, ['kkw','1234',['파이썬','자바']]]  -> 3차원 리스트(배열안에 배열안에 배열)
#                             0       1
#              0      1           2
#      0 1        2
# f변수에서 파이썬을 가지고 오고싶으면 f[2][2][0]읻
```

```python
# 리스트에 C R U D 기법을 적용해보자.
# CREATE -> 생성, 추가, 삽입 .append()
# READ -> 전체 리스트 보기, 1개 자료보기 .index()
# UPDATE -> 자료 수정 passwords[3] = "4486"
# DELETE -> 자료 삭제 del passwords[3] -> 지워짐 or .remove(3) .pop(3) 도 있음 remove와 pop를 많이 사용함.
```

```python
# 리스트에 자료를 추가해보자!
ids = ["kkw", "lhj", "ksb"] #회원 id
ids.append("MBC")
print(ids)
```

```python
passwords = ["1234", "4321", "4567"]
passwords.append("8888")
print(passwords)
```

```python
# 리스트에 있는 값 가져오기
ids.index("MBC")
```

```python
# MBC ID의 인덱스를 알아봐 패스워드의 값을 찾아온다.
passwords[3]
```

```python
# 알아온 password 값을 4486으로 수정해본다.
passwords[3] = "4486"
print(passwords)
```

```python
# 3번째 인덱스에 암호를 삭제해본다.
passwords.remove("4486")
print(passwords)
```

```python
# pop은 삭제되는 내용을 출력하고 삭제한다.
# passwords.pop() -> 맨 마지막 자료 삭제
# passwords.pop(2) -> 2번 인덱스를 삭제
passwords.pop(2)
```
