```python
# 함수는 반복적인 코드를 위쪽에 선언하여 호출함으로써
# 프로그램을 가독성 좋고 간편하게 구현할 수 있다.

# 클래스가 없는 기능을 함수라고 하고
# 클래스가 있는 기능을 메서드라고 한다.

# def는 함수나 메서드를 만드는 예약어로
# def 함수명(매개변수 == 인수 == 파라미터) :
#                                    들여쓰기 필수
#     함수 내부의 실행문들
#     return 값 함수 종료시 호출한 곳으로 넘겨주는 값.
# 들여쓰기를 끝내면 함수 종료

# 변수 = 함수명(매개변수 == 인수 == 파라미터)
# 위에서 만든 함수를 호출하여 처리 후 리턴을 제공받는다.
```

```python
# 함수를 만든다.(함수 정의)
def add(a,b):   # add() 함수를 만들고 인수는 a와 b를 받는다.
    return a+b   # 받은 a, b를 더하고 돌려준다.
```

```python
# 만약 입력값이 몇개가 될지 모를때는?
def add_many(*args):   # *을 붙이면 몇개의 값이 들어오든! 이라는 뜻
    # *의 뜻은 와일드 문자로 김씨로 시작한다.~ "김*"
    # "*원" 원으로 끝나는 문자
    # "*딸기*" 글자에 '딸기'가 포함된다는 의미
    # args는 인수, 숫자일때 많이 씀(들어오는 수)
    # args는 인수라는 뜻으로 다른 글자를 사용해도 됨.
    # 그러나 프로그래머들은 관례로 대부분 args로 사용함.
    result = ""   # 결과를 보관하는 변수
    for i in args:   # args는 튜플(불변값)로 들어옴.
        result = result + i   # 튜플로 되어있는 값을 번갈아 더함.
    return result   # 들여쓰기 잘보기!!!!!
```

```python
# add_many 테스트
result = add_many(1,2,3)
print(result)
```

```python
# add_many 테스트2
result = add_many(1,2,3,6)
print(result)
```

```python
resultText = add_many("안녕하세요","반갑습니다","파이썬 지옥입니다.")
print(resultText)
# 튜플은 숫자형인데 문자열을 넣었기 때문에 오류남
# 문자열은 왠만하면 쓰지말기
```

```python
# 함수에 초기값 지정하기
def say_myself(name, age, man=True):
    #                       man은 기본값으로 true 지정
    print(f"나의 이름은 {name}입니다.")
    print(f"나의 나이는 {age}세 입니다.")
    if man :
        print("저는 남자입니다.")
    else:
        print("저는 여자입니다.")
```

```python
# 변수 선언에는 전역변수가 있고 지역변수가 있다.
# 전역변수는 함수나 if, for, while문 바깥쪽(맨위)에 작성하는 것이 좋다.
# 지역변수는 함수나 if, for, while문에서 1회용으로 사용되는 것으로 안쪽에 선언.
```

```python
g1 = 1
def vartest(g1):
    g1 = g1 + 1

vartest(g1)
print(g1)   # 값이 1인 이유는 전역변수로 g1을 지정했기 때문에 def함수가 실행이 안됨.
```

```python
g1 = 1
def vartest(g1):
    g1 = g1 + 1
    return g1   # return 함수를 사용하면 결과값이 틀려짐.
# 변수에 return을 넣으면 받을 수 있는 형태가 됨.
g1 = vartest(g1)   # 함수 실행 후 결과를 g1에 다시 넣음.
print(g1)
```

```python
def vartest1(a):
    a = a + 1
    #return 없음
vartest1(3)
print(a)
# a라는 변수가 어디에도 정의되지 않았기 때문에 오류 발생
# a라는 변수는 로컬 변수임으로 def가 종료되면 사라진다.
```

```python
a = None   # def밖에 있는 a와 (전역변수a)
def vartest2(a):   # def안에있는 a와 틀리기 때문에(지역변수a)
    # a변수를 안에 써줘야 def가 작동하고 vartest2가 실행됨.
    # 결론은 서로 다른 a변수.
    a = a + 1
    # 리턴없음
    # 하지만 리턴이 없기 때문에 그냥 None이 나옴.
vartest2(3)
print(a)
```

```python
a = None
def vartest3(a):
    a = a + 1
    return a   #위 코드와 변화된 내용은 return과
a = vartest3(3)   # 함수 처리 후 결과를 a에 넣음.
# a는 def함수 실행 후 vartest3 결과값 도출
# JAVA 에서는 this 함수랑 같은 의미임.
print(a)
```

```python
# return 없이 글로번 변수를 사용하고 싶은 경우
session = "로그인안함"   #문자열처리
def login(id):
    """
    키보드로 입력받은 문자열을
    리스트에서 찾아 중복 검사를 진행하고
    없으면 회원가입을 할 수 있도록 처리

    작성자 : 김기원
    :parameter id : 문자열
    :returns : 로그인 상태를 세션에 저장한다.(글로벌)
    """
    global session    #글로벌 명령어는 전역변수를 가져옴.
    session = id   #글로벌 명령어로 같은 변수명은 사용 금지

login(input("아이디를 입력하세요 : "))
print(login.__doc__)   #독스트링
```
